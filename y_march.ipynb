```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("US_youtube_trending_data.csv")
```

```python
df.head()
```

```python
df3 = df[["title","trending_date","channelTitle","categoryId","tags","view_count","likes","dislikes","comment_count","thumbnail_link","description"]].copy()
df3 = df3[df3["likes"] != 0]
df3 = df3[df3["view_count"] != 0]
df3 = df3[df3["tags"] != "[None]"]

df3["trending_date"] = pd.to_datetime(df3["trending_date"])
df3 = df3[(df3["trending_date"].dt.year == 2023)]

df3["best_ratio_1"] = ((df3["likes"]) / df3["view_count"]) + (df3["comment_count"] / df3["view_count"])
df_top10 = df3.nlargest(20, "best_ratio_1")
df_top10
```

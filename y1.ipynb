```python
df2 = df[["title","channelTitle","categoryId","tags","view_count","likes","dislikes","comment_count","thumbnail_link","description"]].copy()
df2 = df2[df2["likes"] != 0]
df2 = df2[df2["view_count"] != 0]
df2 = df2[df2["dislikes"] != 0]
df2 = df2[df2["tags"] != "[None]"]
df2["ratio"] = (df2["likes"] - df2["dislikes"])/df2["view_count"]
```

```python
df2.head()
```

```python
df_top10 = df2.nlargest(20, "ratio")
df_top10
```

```python
df3 = df[["title","trending_date","channelTitle","categoryId","tags","view_count","likes","dislikes","comment_count","thumbnail_link","description"]].copy()
df3 = df3[df3["likes"] != 0]
df3 = df3[df3["view_count"] != 0]
df3 = df3[df3["dislikes"] != 0]
df3 = df3[df3["tags"] != "[None]"]

df3["trending_date"] = pd.to_datetime(df3["trending_date"])
df3 = df3[(df3["trending_date"].dt.year == 2021)]

df3["best_ratio_1"] = ((df3["likes"] - df3["dislikes"]) / df3["view_count"]) + (df3["comment_count"] / df3["view_count"]) * 30
df_top10 = df3.nlargest(20, "best_ratio_1")
df_top10
```

```python
df4 = df_top10
tags_df = df4['tags'].str.get_dummies(sep='|')
df_with_tags_matrix = pd.concat([df4, tags_df], axis=1)
```
